## Regression Tool - 1

```python
from sklearn.datasets import load_linnerud
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset with pandas DataFrames
data = load_linnerud(as_frame=True)

# Features (exercise performance)
X = data.data

# Targets (physiological measurements)
y = data.target

# Here, X contains Situps, Chins, and Jumps, while y contains Weight, Waist, and Pulse.

# Combine into a single DataFrame if needed
df = pd.concat([X, y], axis=1)

# View the first few rows
print(df.head())
```

```python
# Creating a Regression Model: Waistline vs. Situps
situps = df['Situps'].values.reshape(-1, 1)
waistline = df['Waist'].values

# Fit the linear regression model
mymodel = LinearRegression()
mymodel.fit(situps, waistline)
```

```python
# Predicting Waistline from Situps
predicted_waistline = mymodel.predict(situps)

# Plotting the results
plt.scatter(situps, waistline, color='blue', label='Actual Waistline')
plt.plot(situps, predicted_waistline, color='red', label='Predicted Waistline')
plt.xlabel('Situps')
plt.ylabel('Waistline')
plt.title('Regression Model: Waistline vs. Situps')
plt.legend()
plt.show()
```

### Fit the model for Weight vs. Situps

```python
# Fit the model for Weight vs. Situps
situps = df['Situps'].values.reshape(-1, 1)
weight = y['Weight']

# Fit the linear regression model for Weight
mymodel = LinearRegression()
mymodel.fit(situps, weight)

# Predicting Weight from Situps
predicted_weight = mymodel.predict(situps)

# Plotting the results for Weight
plt.scatter(situps, weight, color='green', label='Actual Weight')
plt.plot(situps, predicted_weight, color='orange', label='Predicted Weight')
plt.xlabel('Situps')
plt.ylabel('Weight')
plt.title('Regression Model: Weight vs. Situps')
plt.legend()
plt.show()
```

### Fit the model for Pulse vs. Situps


```python
# Fit the model for Pulse vs. Situps
situps = df['Situps'].values.reshape(-1, 1)


# Fit the linear regression model for Pulse
pulse = y['Pulse']
mymodel.fit(situps, pulse)

# Predicting Pulse from Situps
predicted_pulse = mymodel.predict(situps)


# Plotting the results for Pulse
plt.scatter(situps, pulse, color='purple', label='Actual Pulse')
plt.plot(situps, predicted_pulse, color='brown', label='Predicted Pulse')
plt.xlabel('Situps')
plt.ylabel('Pulse')
plt.title('Regression Model: Pulse vs. Situps')
plt.legend()
plt.show()

```

### Fit the model for Chins vs. Waist


```python
# Fit the model for Chins vs. Waist
waist = y['Waist']
chins = X[['Chins']]


# Fit the linear regression model for Chins
mymodel = LinearRegression()
mymodel.fit(chins, waist)

# Predicting Waist from Chins
predicted_waist_from_chins = mymodel.predict(chins)


# Plotting the results for Waist vs. Chins
plt.scatter(chins, waist, color='cyan', label='Actual Waist from Chins')
plt.plot(chins, predicted_waist_from_chins, color='magenta', label='Predicted Waist from Chins')
plt.xlabel('Chins')
plt.ylabel('Waist')
plt.title('Regression Model: Waist vs. Chins')
plt.legend()
plt.show()
```

```python
# Repeating thesame fitting and plotting for

mymodel.fit(chins, weight)
mymodel.fit(chins, pulse)

# Define the Jumps feature and fit the model for Waist, Weight, and Pulse

pairs = [
    ('Chins', 'Weight'),
    ('Chins', 'Pulse'),
    ('Jumps', 'Waist'),
    ('Jumps', 'Weight'),
    ('Jumps', 'Pulse')
]

for feature, target in pairs:
    feature_data = X[[feature]].values.reshape(-1, 1)
    target_data = y[target].values
    mymodel.fit(feature_data, target_data)
    predicted_target = mymodel.predict(feature_data)
    plt.scatter(feature_data, target_data, label=f'Actual {target}')
    plt.plot(feature_data, predicted_target, label=f'Predicted {target}')
    plt.xlabel(feature)
    plt.ylabel(target)
    plt.title(f'Regression Model: {target} vs. {feature}')
    plt.legend()
    plt.show()
```
