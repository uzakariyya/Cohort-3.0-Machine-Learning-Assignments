```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
path = r"C:\Users\LENOVO\Cohort-3.0-Machine-Learning-Assignments\data\penguins_lter.csv"

# Show first few rows
df = pd.read_csv(path)
print(df.head())
print(df.info())
```

# Data Cleaning

```python
# Drop missing values for simplicity
df.dropna(inplace=True)

# Basic info
df.info()
```

ðŸ“Š Histogram 1: Flipper Length Distribution



```python
plt.figure(figsize=(8,5))
sns.histplot(df['Flipper Length (mm)'], bins=20, kde=True)
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Count')
plt.show()
```

ðŸ“Š Histogram 2: Body Mass by Species

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Body Mass (g)', hue='Species', bins=20, kde=True)
plt.title('Body Mass Distribution by Species')
plt.xlabel('Body Mass (g)')
plt.ylabel('Count')
plt.show()
```

ðŸ“Š Histogram 3: Culmen Depth

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Culmen Depth (mm)'], bins=15, color='orange')
plt.title('Distribution of Culmen Depth')
plt.xlabel('Culmen Depth (mm)')
plt.ylabel('Count')
plt.show()
```

ðŸ“Š Histogram 4: Flipper Length by Sex

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Flipper Length (mm)', hue='Sex', bins=20, kde=True)
plt.title('Flipper Length Distribution by Sex')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Count')
plt.show()
```

ðŸ“Š Histogram 5: Island Distribution

```python
plt.figure(figsize=(7,5))
sns.histplot(data=df, x='Island', hue='Species', multiple='stack')
plt.title('Penguin Species Count by Island')
plt.xlabel('Island')
plt.ylabel('Count')
plt.show()
```
