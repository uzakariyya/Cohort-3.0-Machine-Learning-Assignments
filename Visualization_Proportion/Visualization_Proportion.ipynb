# Pie chart

Create a pie chart displaying the proportion of Poisonous vs. Edible mushrooms

```python
import matplotlib.pyplot as plt

# Sample data: mushroom classes
labels = ['Edible', 'Poisonous']
counts = [4208, 3916]
colors = ["#0080ff", "#ff0000"]


```

```python
# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Proportion of Edible vs Poisonous Mushrooms')
plt.axis('equal')
plt.show()
```

```python
capcolor = mushrooms.groupby(['cap-color']).count()
capcolor
```

```python
habitat = mushrooms.groupby(['habitat']).count()
habitat
```

```python
habitat_counts = mushrooms['habitat'].value_counts()
```

```python
# Labels and values
labels = habitat_counts.index
sizes = habitat_counts.values
colors = plt.cm.Paired.colors

# Pie (Donut) Chart
plt.figure(figsize=(7, 7))
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Draw center circle to make it a donut
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of Mushrooms by Habitat')
plt.tight_layout()
plt.show()
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
import pandas as pd

# Example data (replace this with your actual cap-color counts)
data = {
    'color': ['brown', 'buff', 'cinnamon', 'green', 'pink', 'purple', 'red', 'white', 'yellow'],
    'amount': [2284, 1616, 400, 75, 100, 200, 300, 2000, 576]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=90,  
    values=df.amount,
    labels=[f"{label} ({val})" for label, val in zip(df.color, df.amount)],
    figsize=(30, 30),
    colors=["brown", "tan", "maroon", "green", "pink", "purple", "red", "whitesmoke", "yellow"],
    title={'label': 'Distribution of Mushroom Cap Colors', 'loc': 'center', 'fontsize': 24},
    legend={'loc': 'upper right', 'bbox_to_anchor': (1.1, 1.0), 'fontsize': 12, 'title': 'Cap Colors'   },
)
plt.show();
```
