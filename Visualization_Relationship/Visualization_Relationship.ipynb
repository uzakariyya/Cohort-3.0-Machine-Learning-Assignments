# Visualization Relationship

```python
import matplotlib.pyplot as plt
import pandas as pd     
import seaborn as sns  
```

```python
# Load the dataset
honey = r"C:\Users\LENOVO\Cohort-3.0-Machine-Learning-Assignments\data\honey.csv"
df = pd.read_csv(honey)
# Display the first few rows of the dataset
df.head()

# Convert column names to lowercase and strip whitespace
df.columns = df.columns.str.lower().str.strip()

# Display the columns, info, and descriptive statistics of the DataFrame
df.columns



```

#### Total Honey Production Over Time

```python

#plotting the total honey production over the years


yearly_prod = df.groupby('year')['totalprod'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_prod, x='year', y='totalprod', marker='o')
plt.title("Total Honey Production Over Time (National Trend)")
plt.ylabel("Total Production (lbs)")
plt.xlabel("Year")
plt.grid()
plt.tight_layout()
plt.show()
```

### Total Production by State and Year (Heatmap)

```python
# Total Honey Production Over Time
# Displaying the total honey production over time using a line plot


prod_heatmap = df.groupby(['state', 'year'])['totalprod'].sum().unstack()

plt.figure(figsize=(14, 10))
sns.heatmap(prod_heatmap, cmap="YlGnBu", linewidths=0.3)
plt.title("Total Honey Production by State and Year")
plt.xlabel("Year")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

### Top 5 States by Total Production Trend

```python
# Top 5 States with Highest Honey Production Over Time
# Filtering the top 5 states based on total production and plotting their yearly production trends

top_states = df.groupby('state')['totalprod'].sum().nlargest(5).index
top_df = df[df['state'].isin(top_states)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=top_df, x='year', y='totalprod', hue='state', marker='o')
plt.title("Yearly Honey Production in Top 5 States")
plt.xlabel("Year")
plt.ylabel("Total Honey Production (lbs)")
plt.grid(True)
plt.legend(title="State")
plt.show()
```
